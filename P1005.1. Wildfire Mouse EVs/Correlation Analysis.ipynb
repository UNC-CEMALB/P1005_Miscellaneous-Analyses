# Spearman Correlations

```r
# getting dfs for saline or red oak only
saline_df = combined_df %>%
    filter(Exposure == "Saline")
red_oak_df = combined_df %>%
    filter(Exposure == "RedOakFlame")

# getting a df for unique combinations to iterate through
unique_molecules = combined_df %>%
    select(miRNA, mRNA) %>%
    unique()
```

```r
both_correlation_df = data.frame()
saline_correlation_df = data.frame()
red_oak_correlation_df = data.frame()

# creating a function to run spearman tests
get_spearman_correlations = function(df, corr_df, estimate, pvalue, padjust){
    for (i in 1:length(unique_molecules$miRNA)){
        exposure_df = df %>%
            filter(miRNA == unique_molecules$miRNA[i] & mRNA == unique_molecules$mRNA[i]) 

        spearman_test = cor.test(exposure_df$miRNA_Value, exposure_df$mRNA_Value, method = c('spearman'))

        # remembering values of interest
        values_df = data.frame(unique_molecules$miRNA[i], unique_molecules$mRNA[i], spearman_test$estimate, 
                               spearman_test$p.value)
        corr_df = rbind(corr_df, values_df)
    }

    # dropping duplicate rows
    corr_df = corr_df %>% 
        unique()

    # adding col names
    colnames(corr_df) = c("miRNA", "mRNA", estimate, pvalue)
    
    # combining miRNA & mRNA columns to make formatting easier later
    corr_df = corr_df %>%
        unite("miRNA_mRNA", miRNA:mRNA, sep = "_", remove = TRUE)

    # calculating padj values
    corr_df[[padjust]] = p.adjust(as.numeric(as.character(corr_df[[pvalue]])), method = "fdr")
    
    return(corr_df)
}
```

```r
# calling function
both_correlation_df = get_spearman_correlations(combined_df, both_correlation_df, "Both_Estimate", "Both_P_Value",
                                                "Both_P_Adjust")
saline_correlation_df = get_spearman_correlations(saline_df, saline_correlation_df,  "Saline_Estimate", 
                                                  "Saline_P_Value", "Saline_P_Adjust")
red_oak_correlation_df = get_spearman_correlations(red_oak_df, red_oak_correlation_df, "Red_Oak_Flame_Estimate", 
                                                   "Red_Oak_Flame_P_Value", "Red_Oak_Flame_P_Adjust")
```

```r
# combining into 1 df
correlation_df = full_join(full_join(both_correlation_df, saline_correlation_df), red_oak_correlation_df)
head(correlation_df)
```

```r
# joining with the appendix df
final_df = full_join(appendix_df %>%
         rename(miRNA_mRNA = Temp.ID.that.could.help.with.Merging), correlation_df)

head(final_df)
```
